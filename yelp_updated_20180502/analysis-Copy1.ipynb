# Logistic regression modeling and interpretation using Yelp reviews

```python
%matplotlib inline
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
reviews = pd.read_csv('data/reviews.csv')
```

## <span style="color:blue">   Logistic regression Run #1

### Feature inportance will be checked with scaled input data

# Normalize data

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))

y = reviews['regular']
X = scaler.fit_transform(reviews.drop(columns=['regular']))
X = pd.DataFrame(X, columns=reviews.drop(columns=['regular']).columns)
```

# Fit a linear model

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=0)
clf.fit(X, y)
```

# Analysis on coefficients

```python
pd.Series(clf.coef_[0], index=X.columns).plot(kind='bar');
```

## <span style="color:blue">  Logistic Regression Run #2
